```python
from bs4 import BeautifulSoup
import requests

# url = '''
#     https://kr.hotels.com/search.do?resolved-location=NEIGHBORHOOD%3A1747961%3AUNKNOWN%3AUNKNOWN&destination-id=1747961&q-destination=%EC%97%AC%EC%9D%98%EB%8F%84,%20%EC%84%9C%EC%9A%B8,%20%ED%95%9C%EA%B5%AD
#     &q-check-in=2019-12-25&q-check-out=2019-12-26&q-rooms=1&q-room-0-adults=2&q-room-0-children=0"
#     '''

url = "https://kr.hotels.com/search.do?resolved-location=NEIGHBORHOOD%3A1747961%3AUNKNOWN%3AUNKNOWN&destination-id=1747961&q-destination=%EC%97%AC%EC%9D%98%EB%8F%84,%20%EC%84%9C%EC%9A%B8,%20%ED%95%9C%EA%B5%AD&q-check-in=2019-12-25&q-check-out=2019-12-26&q-rooms=1&q-room-0-adults=2&q-room-0-children=0"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
print(soup)

prices = soup.find_all('div', class_="price")
len(prices)
for price in prices:
    print(price.text)
```
